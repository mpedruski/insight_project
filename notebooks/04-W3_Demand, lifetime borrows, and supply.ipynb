```python
### Load up the files and packages

from joblib import dump
import numpy as np
import pandas as pd
from pathlib import Path
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
from sklearn.metrics import mean_squared_error


### Determine paths to datasets and load into pandas dataframe
cleaned_folder = Path("../data/cleaned")
file_to_open = cleaned_folder / 'by_titles.csv'
df = pd.read_csv(file_to_open)

sns.set()

```

```python
sns.scatterplot(df['Lifetime'],df['Demanded'])
plt.savefig('./figures/Lifetime_current.png')
```

```python
sns.scatterplot(df['Total'],df['Demanded'])
plt.plot([0, 250], [0,250], linewidth=2)
plt.savefig('./figures/Total_current.png')
```

```python
lm = linear_model.LinearRegression()
lm.fit(np.array(df['Total']).reshape(-1,1),df['Demanded'])
lm.score(np.array(df['Total']).reshape(-1,1),df['Demanded'])
```

```python
mean_squared_error(df['Total'], df['Demanded'], squared = False)/np.mean(df['Total'])
```
